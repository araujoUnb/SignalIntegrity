```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Time array (x-axis)
time = np.linspace(0, 10e-9, 1000)

# Good Signal Integrity (ideal waveform)
good_signal = np.where(time < 5e-9, 0, 3.3)

# Poor Signal Integrity (distorted waveform)
poor_signal = 3.3 * (1 - np.exp(-time / 1e-9))
poor_signal = np.where(time < 5e-9, poor_signal, 3.3 - poor_signal)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, good_signal, label='Good Signal Integrity')
plt.plot(time, poor_signal, label='Poor Signal Integrity')
plt.xlabel('Time (ns)')
plt.ylabel('Voltage (V)')
plt.legend()
plt.title('Signal Integrity')
plt.grid(True)
plt.show()
```
